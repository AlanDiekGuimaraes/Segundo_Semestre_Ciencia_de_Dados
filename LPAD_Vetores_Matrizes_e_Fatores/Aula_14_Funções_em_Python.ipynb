```python
import numpy as np
import matplotlib.pyplot as plt
import math

# Usando a lib math
# Cosseno de 60 graus
print(math.cos(math.radians(60)))
# Cossendo de pi/radianos
print(math.cos(math.pi/3))

# Usando a lib numpy
# Cosseno de 60 graus
print(np.cos(np.radians(60)))
# Cossendo de pi/radianos
print(np.cos(np.pi/3))
```

```python
fcos = np.cos(np.linspace(0,360,360) * np.pi/180)
x = np.linspace(0,360,360)

plt.figure().set_figheight(2,3)
plt.plot(x, fcos)
plt.xlabel('x', fontsize=15)
plt.ylabel('cos(x)', fontsize=15)
plt.title('Função Cosseno')
plt.legend(['f(x)=cos(x)'])
plt.show()
```

```python
x = np.linspace(0,360,360)
fcos = np.cos(2*(x * np.pi/180)) - 3

plt.figure().set_figheight(2,4)
plt.plot(x, fcos)
plt.xlabel('x', fontsize=15)
plt.ylabel('cos(x)', fontsize=15)
plt.title('Função Cosseno')
plt.legend(['f(x)= cos(2x) - 3'])
plt.show()
```

```python
# Usando a lib math
# Tangente de 60 graus
print(math.tan(math.radians(60)))
# Tangente de pi/radianos
print(math.tan(math.pi/3))

# Usando a lib numpy
# Cosseno de 60 graus
print(np.tan(np.radians(60)))
# Tangente de pi/radianos
print(np.tan(np.pi/3))
```

```python
x = np.linspace(0,360,360)
ftan = np.tan(x * np.pi/180)

plt.figure().set_figheight(2,3)
plt.plot(x, ftan)
plt.xlabel('x', fontsize=15)
plt.ylabel('tab(x)', fontsize=15)
plt.title('Função Tangente')
plt.legend(['f(x) = tan(x)'])
plt.ylim(-20, 20)
plt.show()
```

```python
x = np.linspace(-100,100)
y = x**3 + 3*x + 5

plt.plot(x, y)
plt.xlabel('x', fontsize=15)
plt.ylabel('y', fontsize=15)
plt.title('Função Polinomial')
plt.legend(['f(x) = x³ + 3x + 5'])
plt.show()
```

```python
x = np.linspace(0.1, 10, 30)
y = 2.**x

plt.plot(x, y)
plt.xlabel('x', fontsize=15)
plt.ylabel('y', fontsize=15)
plt.title('Função Exponencial')
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.show()
```

## Função log no python
Para obter o logaritimo de um número, utilize o comando math.log(b, base)

```python
import math
x = math.log(16,2)
print(f'O log de 16 na base 2 é {x}')
```

```python
import numpy as np
import math
import matplotlib.pyplot as plt

x = np.linspace(0.1, 10, 30)
y = np.log(x)
plt.xlabel('x', fontsize=15)
plt.ylabel('y', fontsize=15)
plt.title('Função logarítmica')
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.plot(x, y)
```
