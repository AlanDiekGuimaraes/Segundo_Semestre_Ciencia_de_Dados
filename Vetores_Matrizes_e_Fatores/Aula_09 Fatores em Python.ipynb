Para gerar os valores de x, utilize o comando numpy.array()

```python
#Função Linear
import numpy as np
#Variáveis
x = np.array([-2, -1, 0, 1, 2])

y=2*x
print(y)
```

Para plotar a função, utilize a biblioteca matplotlib.pyplot()

```python
#Gráfico
import matplotlib.pyplot as plt
import numpy as np

# plt.figure().set_figheight(3. 4)
plt.plot(x, y)
plt.xlabel('X', fontsize=15)
plt.ylabel('Y', fontsize = 15)
plt.title('Função Linear')
plt.legend(['y=2x'])
plt.show()
```

Um usuário fornecerá dois pontos pertencentes a uma reta, crie o código que exibirá a lei de formação da reta e exibirá o grafico dessa reta. 

```python
xA = float(input('Digite xA: '))
yA = float(input('Digite yA: '))
xB = float(input('Digite xB: '))
yB = float(input('Digite yB: '))

try:
  a = (yA - yB) / (xA - xB)
  b = yA - a*xA

  x = np.array([-2, -1, 0, 1, 2])
  y = a*x + b
  print(f'Valor de a: {a}')
  print(f'Valor de b: {b}')
  print(f'Lei de formação: y = {a}x + {b}')

  plt.plot(x, y)
  plt.xlabel('x', fontsize = 15)
  plt.ylabel('y', fontsize = 15)
  plt.title('Função Linear')
  plt.legend([f'y = {a}x + {b}'])
  plt.grid()
  plt.show()

except:
  print('Deu erro na conta aqui kkkkk.')
```

```python
def leiFormação(x1, y1, x2, y2, x):
  A = (y2 - y1) / (x2 - x1)
  B = (A * x2) - y2
  y = A*x + B
  return y


x = np.array([-2, -1, 0, 1, 2])

y = leiFormação(-2, 0, 4, 12, x)
print(y)

plt.plot(x, y)
plt.xlabel('x', fontsize = 15)
plt.ylabel('y', fontsize = 15)
plt.title('Função Linear')
plt.legend([f'y = {a}x + {b}'])
plt.grid()
plt.show()
```

## Subplot no Python
Para criar vários gráficos em uma única figura, utilize o comando matplotlib.pyplot.subplot()

```python
import matplotlib.pyplot as plt
import numpy as np

# Plot 1
# plt.figure() .set_figheight(2,3)
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])
plt.subplot(1, 2, 1)
plt.xlabel('x', fontsize = 15)
plt.ylabel('y', fontsize = 15)
plt.title('Serra')
plt.plot(x, y)

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])
plt.subplot(1, 2, 2)
plt.xlabel('x', fontsize = 15)
plt.ylabel("Y", fontsize=15)
plt.title("Linear")
plt.plot(x,y)
plt.tight_layout(pad=5.0)
plt.show()
```

```python
x = np.arange(-10, 10, 1)
y1 = 2*x + 6
y2 = -3*x + 1
y3 = x - (1/5)
y4 = -6*x - 2

plt.subplot(2, 2, 1)
plt.plot(x, y1)
plt.title('y = 2x + 6')

plt.subplot(2, 2, 2)
plt.plot(x, y2)
plt.title('y = -3x + 1')

plt.subplot(2, 2, 3)
plt.plot(x, y3)
plt.title('y = x - (1/5)')

plt.subplot(2, 2, 4)
plt.plot(x, y4)
plt.title('y = -6x - 2')

plt.show()
```
